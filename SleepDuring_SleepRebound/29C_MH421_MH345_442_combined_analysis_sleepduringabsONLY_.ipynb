```python
import scipy
from scipy import signal, misc
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import matplotlib
```

```python
ds=pd.read_csv('MH421_442R2_combinedRawSleepData.txt', delimiter='\t')
dst=ds.T #transpose
dst
```

```python
dst.head(20)
```

```python
#calculate day2 night sleep (ZT12-24, or Day2.5-3) for sleep during
#dst[2161] is day 2.5
#dst[2881] is day 3
day2nightsleep=dst.loc[:,2162:2881]
day2nightsleep.head(20)
```

```python
day2nightsleep = day2nightsleep.dropna() #gets rid of the empty columns
type(day2nightsleep[2163][0])
```

```python
day2nightsleep.sum(axis=1) #sums the columns selected
```

```python
day2nightsleep_sum=day2nightsleep.sum(axis=1) #sums the columns selected
type(day2nightsleep_sum)
```

```python
day2nightsleep_sum=pd.DataFrame(day2nightsleep_sum)
day2nightsleep_sum=day2nightsleep_sum.drop(day2nightsleep_sum.index[0]) #drop first row
day2nightsleep_sum.columns= ['Sleep_Night2']
day2nightsleep_sum["Genotype"]=day2nightsleep_sum.index
day2nightsleep_sum['Genotype']=day2nightsleep_sum['Genotype'].map(lambda x: x.rstrip('.txt').rstrip('.1234567890$MCtHR')) #use map to replace with new value
day2nightsleep_sum
```

```python
day2nightsleep_sum['Sleep_Night2']=pd.to_numeric(day2nightsleep_sum['Sleep_Night2']) #converts to numeric (float)
```

```python
day2nightsleep_sum=day2nightsleep_sum.reset_index()
day2nightsleep_sum=day2nightsleep_sum.drop(['index'], axis=1)
day2nightsleep_sum
```
