```python
#calculate day3 daytime sleep (ZT0-6, or Day 3-3.25) for sleep during
#dst[2882] is day 3
#dst[3241] is day 3.25
day3ZT0to6sleep=dst.loc[:,2882:3241]
day3ZT0to6sleep.head(20)
```

```python
day3ZT0to6sleep = day3ZT0to6sleep.dropna() #gets rid of the empty columns
day3ZT0to6sleep
```

```python
day3ZT0to6sleep.sum(axis=1) #sums the columns selected
```

```python
day3ZT0to6sleep_sum=day3ZT0to6sleep.sum(axis=1) #sums the columns selected
day3ZT0to6sleep_sum=day3ZT0to6sleep_sum.drop(day3ZT0to6sleep_sum.index[0]) #drop first row
day3ZT0to6sleep_sum.head(20)
```

```python
day3ZT0to6sleep_sum=pd.DataFrame(day3ZT0to6sleep_sum)
day3ZT0to6sleep_sum.columns= ['Day3ZT0-6']
day3ZT0to6sleep_sum['Genotype']=day3ZT0to6sleep_sum.index
```

```python
day3ZT0to6sleep_sum['Genotype']=day3ZT0to6sleep_sum['Genotype'].map(lambda x: x.rstrip('.txt').rstrip('.1234567890$MCtHR')) #use map to replace with new value
day3ZT0to6sleep_sum=day3ZT0to6sleep_sum.reset_index()
day3ZT0to6sleep_sum=day3ZT0to6sleep_sum.drop(['index'], axis=1)
day3ZT0to6sleep_sum
```

```python
day3ZT0to6sleep_sum['Day3ZT0-6']=pd.to_numeric(day3ZT0to6sleep_sum['Day3ZT0-6']) #converts to numeric (float)
```

```python
day3ZT0to6sleep_sum.groupby('Genotype')[['Day3ZT0-6']].median()
```
